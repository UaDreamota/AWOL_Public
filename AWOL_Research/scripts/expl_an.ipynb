```python
slin['age'] = pd.to_datetime(slin['birth_date'], errors='coerce').apply(lambda x: pd.Timestamp.now() - x).dt.days // 365
nd = slin.drop_duplicates(['detail_url'])
```

```python

# Filter ages between 18 and 70
filtered_ages = slin['age']
filtered_ages = filtered_ages[(filtered_ages >= 18) & (filtered_ages <= 70)]

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_ages, bins=53, edgecolor='black')  # 53 bins = 70 - 18 + 1 for 1-year bins

plt.title('Дистрибуція віку якщо скрейпіти UALosses')
plt.xlabel('Вік')
plt.ylabel('Кількість загиблих')
plt.xlim(18, 70)
plt.grid(True)

# Set x-ticks from 18 to 70 every 1 year
ticks = np.arange(18, 71, 1)
plt.xticks(ticks)

plt.tight_layout()
plt.show()
```
